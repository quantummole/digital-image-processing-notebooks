Histogram equalization is a form of distribution transformation of random variables. Where the color distribution is transformed from an arbitrary distribution to a uniform distribution.

```python
#compute histogram
bins = [0]*256
def incBin(x) :
    bins[x]+=1
f = np.vectorize(incBin)
f(im)
plt.plot(bins)
plt.show()
#compute cumulative distribution
Fx = [0]*256
Fx[0] = bins[0]
for i in range(1,256) :
    Fx[i] = Fx[i-1]+bins[i]
normalizer = 1.0/(im.shape[0]*im.shape[1])
Fx = [min(1,x*normalizer) for x in Fx]
#do the transformation
def transform(x) :
    return Fx[x]
hist_eq = np.vectorize(transform)
im_eq = img_as_ubyte(hist_eq(im))
io.imshow(im_eq)
plt.show()
bins = [0]*256
f(im_eq)
plt.plot(bins)
plt.show()
```

we can notice that the final histogram as the intensities distributed throughout the dynamic range.
