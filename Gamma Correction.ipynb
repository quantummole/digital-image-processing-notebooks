Gamma Correction: It is an intensity transformation where the pixel intensities are adjusted using a power law.

```python
gamma = np.log(0.5)/np.log(np.mean(im[:]))
im_corrected = im ** gamma
io.imshow(im_corrected)
plt.show()
```
